## Tools

LLMs have Reasoning (Think) and Language Generation (Speak) capabilities, but does not have ability to execute tasks (Action). They can't, on their own, book airline tickets, access live data and weather, do reliable math, call APIs to make a tweet or send a mail on your behalf, run code, interact with a database, etc.

A Tool is just a Python function (or API) that is packaged in a way the LLM can understand and call when needed.

This is how Tools fit into Agent Ecosystem, and we get Agentic AI. An AI Agent is an LLM-powered ecosystem that can autonomouusly think, decide, and take action using external tools or APIs ot achieve a goal.

Agent = LLM (Reasoning and Decision Making) + Tools (Action)

There are - 
1. Built in Tools
2. Custom Tools

#### How to convert a custom function into a tool, which would then become a runnable and can then interact with LLMs

#### 1. Using @tool 

```python
# Step 1 - create a function

def multiply(a, b):
    """Multiply two numbers"""  # this is just a docstring, but is highly recommended as its more info for LLM later on
    return a*b
```

```python
multiply(3,5) # works now  
```

```python
# Step 2 - add type hints

def multiply(a: int, b:int) -> int:
    """Multiply two numbers"""  
    return a*b
```

```python
multiply(3,5)  # still works
```

```python
# Step 3 - add tool decorator

@tool
def multiply(a: int, b:int) -> int:
    """Multiply two numbers"""
    return a*b
```

Think of @tool like gift wrapping. The gift inside is the same, but the wrapper adds something on top.
For example, @tool takes your plain multiply function and wraps it so that LangChain can recognize it as a tool an agent can use. Your function still multiplies two numbers — the decorator just adds metadata around it.

Essentially, by doing this, a very detailed textual version of the python function goes into the prompt for the LLM, see later.

```python
# multiply(3,5) # doesnt work anymore, TypeError: 'StructuredTool' object is not callable
```

```python
multiply.invoke({"a":3, "b":5})
```

since @tool converted it into a LangChain Tool object, it's no longer a plain function — it's now an object with an .invoke() method. That's why you have to use multiply.invoke({"a": 3, "b": 5}) with a dictionary instead of just multiply(3, 5). 

multiply.invoke({"a":3, "b":5}) is just Langchain executing our Python function locally, no LLMs or API call to OpenAI is being made here.

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
@tool
def muladd(a,b,c):
    """mul a and b"""
    temp = a*b
    """add c"""
    return temp + c
```

```python
muladd.invoke({"a": 1, "b": 2, "c": 3})
```

```python
print(muladd.name)
print(muladd.description)
print(muladd.args)
```

### 2. Using Structured Tool

Special type of tool where the input to the tool follows a structured schema, typically defined using a Pydantic model.

More strictly enforcing constraints.

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
# arg schema using Pydantic
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
def multiply_func(a: int, b: int) -> int:  # int here is not required as such coz already in Pydantic schmea
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="Multiply two numbers",
    args_schema=MultiplyInput  # schema with more constrained types and descriptions of arguments
)
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

### 3. Using BaseTool Class

Base class for all LangChain tools. It defines the core structure and interface that any tool must follow, whether it's a simple one liner or a fully customized function.

This abstract class defines the interface that all LangChain tools must implement. @tool and StructuredTool are built on top on of BaseTool.

Tools are components that can be called by agents to perform specific actions.

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
# arg schema using pydantic

class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"

    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
```

### Toolkits

A toolkit is just a collection (bundle) of related tools that serve a common purpose packaged together for convenience and reusability.

In LangChain, a toolkit might be: GoogleDrive ToolKit

And it can contain the following tools:

1. GoogleDriveCreateFileTool: Upload a file
2. GoogleDriveSearchTool: Search for a file by name/content
3. GoogleDriveReadFileTool: Read contents of a file


Comes in handy bcoz of its reusability, just a make a toolkit and then it can be used in one application and also in another similar kind of application requiring the same tools.

```python
from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add, multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```

```python
tools[0].invoke({"a": 2, "b": 3})
```

```python
tools[1].invoke({"a": 2, "b": 3})
```
